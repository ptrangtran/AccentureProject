```python
import pandas as pd
```

```python
file1 = pd.read_csv('Content.csv')
```

```python
file1.head(10)
```

```python
df_content = file1.dropna()
df_content = df_content.drop(columns=['Unnamed: 0', 'User ID'])
df_content = df_content.rename(columns={'Type' : 'Content Type'})
```

```python
df_content
```

```python
file2 = pd.read_csv('Reactions.csv')
```

```python
df_reactions = file2.dropna()
df_reactions = df_reactions.drop(columns=['Unnamed: 0', 'User ID'])
```

```python
df_reactions
```

```python
file3 = pd.read_csv('ReactionTypes.csv')
```

```python
df_reactiontypes = file3.dropna()
df_reactiontypes = df_reactiontypes.drop(columns=['Unnamed: 0'])
df_reactiontypes
```

```python
df_merged = pd.merge(df_reactions,df_reactiontypes, on= 'Type', how='left')
```

```python
df_merged = pd.merge(df_merged, df_content, on='Content ID', how = 'left')
```

```python
df_merged.info()
```

```python
category_scores = df_merged.groupby('Category')['Score'].sum()
```

```python
df_scores = pd.DataFrame(category_scores)
```

```python
df_scores= df_scores.sort_values(by ='Score', ascending = False)
```

```python
df_scores.head(5)
```

```python
# List of categories to filter
categories_to_select = ['travel', 'science', 'healthy eating', 'animals', 'cooking']

# Filter rows based on selected categories
df_filtered = df_merged[df_merged['Category'].isin(categories_to_select)]
```

```python
df_filtered
```

```python
df_filtered.to_csv('Top5_SocialBuzz.csv', index = False)
```
